# Exercise: Full Eliza Application

Create an interactive chat-bot type program that interacts with the user and has the affect of a simple therapist.
Your program will need to evaluate what the user asks and turn the user's input into a question that sounds like the therapist really cares.

Your first task is to develop a program with a running loop. 
- If the user types in "I am feeling great" or enters "Q", the program will stop running. 
- Your program should print out the last input (as an output) every time before accepting new input.
- Make sure you are accommodating for NO case-sensitivity. (Q is the same as q)

```python
#hollaman
#20191028
#inclass06 Eliza Exercise

from time import sleep #So we're able to delay elizaBot's responses so they seem more... personalized to the user's input
from random import randint #So we're able to randomly select from a set of stock responses

def print_delayed(msg,sec,*sep):
    sleep(sec)
    if sep:
        print(msg,end=sep[0])
    else:
        print(msg,end=' ')
        
#test
#print(randint(5,10))
#print_delayed("... wait wait, I think it's working",5,', sweet!')
```

```python
def find_replace_w_dict(msg, xdict):
    msg_tokenized = msg.lower().strip().split() #lowercase everything, strip spaces, tokenize
    msg_reconstructed = "" #this'll hold the result that will eventually be returned
    
    #nested for loops that check each word against each key noted in the dictionary and..
    for word in msg_tokenized:
        for key in xdict.keys():
            if word==key.lower(): #if word is a key in our dictionary, then..
                word=xdict[key].lower() #replace it with the key's value
                break
        
        msg_reconstructed=" ".join([msg_reconstructed,word]).strip().capitalize() #reconstruct msg and capitalize
    return msg_reconstructed

#create dictionary for use in personalizing responses (hack)

#first2secondP = {"I":"You","am":"are","We":"Y'all", "me":"you", "us":"y'all", "my":"your", "our":"y'all's", "myself":"yourself", "ourselves":"yourselves"}
#testmsg = "I am thinking that we have an opportunity to pool our final project efforts so that the result is a greater portfolio entry we can all applaud ourselves for.."
#print(find_replace_w_dict(testmsg, first2secondP))
```

```python
#We'll instantiate Eliza as a function
def elizaBot():
    
    ##########################
    #Create greetings (r0)
    ##########################
    phrase_greetings = ["Greetings!","This is Eliza-bot.","What is your malfunction?","What brings you here?", "What is the problem from your viewpoint?"]
    
    ##########################
    #Create sample hedges (r1)
    ##########################
    """
    - Please tell me more
    - Many of my patients tell me the same thing
    - It is getting late, maybe we had better quit
    """
    phrase_hedges = ["Please tell me more", "Many of Eliza-bot's patients have spoken on this", "That sounds very heavy.", "Haba! Not as interesting a topic as Eliza-bot initially estimated", "It is getting late, maybe we had better quit"]
    
    
    ##########################
    #Create filler (r2)
    ##########################
    phrase_status = ["Full Disclosure: Eliza-bot is not very good with this kind of problem", "Eliza-bot is requesting more cloud computation resources"]
    
    ################################
    #Create sample qualifiers (r3)
    ################################
    """
    Why do you say that
    You seem to think that
    So, you are concerned that
    """
    phrase_qualifiers = ["Why do you say that", "You seem to think that", "So, you are concerned that"]
    phrase_qualifiers2a = ["How does it make you feel", "that", "when"]
    phrase_qualifiers2b = ["Can you tell Eliza-bot more about", "why you think", "what you do when"]
    
    ##############################
    #Create sample questions (r4)
    ##############################
    phrase_atrisk = ["How are you feeling right now?", "Are you having thoughts about hurting yourself?"]
    phrase_optout = ["Did you want a reference to some good resources?", "Is there a friend or family member you can talk to?"]
    phrase_optout2 = ["Or alternatively, did you want to talk about", "how you're sleeping these days?", "how much exercise you are getting?", "your concrens about the coming AI revolution?"]
    
    ##########################
    #Create sample outro (r5)
    ##########################
    phrase_ohdear1 = ["To the extent Eliza-bot can imagine stuff","Eliza-bot is super optimistic about your future."]
    phrase_ohdear2 = ["Oh dear!", "Well, Eliza-bot wants you to know that you are loved."]
    
    phrase_outro = ["If you require further assistance, please do consider calling 2-1-1. They offer information and referral to community and social services, including counselling services.", "Also, 211 is free, confidential, multilingual and available 24-hours a day.", "And you do not have to give your name or personal details to get information."]
    phrase_outro2 = ["or try 411", "Eliza-bot hears they're good too."]
    
    phrase_positive = ["Eliza-bot wishes you well, Human", "Good Luck!!! You lovely human, you"]
    
    #################
    #input dialogs
    #################
    inputDialog = "Enter your response here or input 'q' to quit: "
    inputDialog2 = "Enter 'y'/'n' or input 'q' to quit: "
    
    
    #################
    #exit conditions
    #################
    exitcondition1 = "I am feeling great"
    
    #-------------------------------------------------------------------#
    
    
    #############################################################
    #create dictionary for use in personalizing responses (hack)
    #############################################################
    first2secondP = {"I":"You","am":"are","We":"Y'all", "me":"you", "us":"y'all", "my":"your", "our":"y'all's", "myself":"yourself", "ourselves":"yourselves"}

    #a few values that'll help direct responses
    counter = 0 #this will count how many times we've walked through the loop
    uresponse = '' #default of user response
    
    atrisk= 0 #if the user is atrisk then we want to take some action
    mode = 1 #we have two modes, one that drives towards an exit and the other that would loop at the user's discretion
    
    while not uresponse == 'q':
        if uresponse == 'i am feeling great':
            break
            
        if counter == 0:
            #Execute Slow Introduction
            print_delayed(phrase_greetings[0],1)
            print_delayed(phrase_greetings[1],2)
            print_delayed(phrase_greetings[randint(2,4)],2,'\n')
            counter+=1
            
        elif counter>=1: #executed after iteration [0], when the user has already provided an input
            if mode == 1: #for the first set of eliza-bot responses, we want to access a limited set of potential replies
                rid = randint(1,3)
            elif mode == 2:
                rid = randint(2,4)
                
            
            ################################
            #Mode 1: qualifiers (r3)
            ################################
            if rid == 3:
                rid2 = randint(0,2)
                uresponsex = find_replace_w_dict(uresponse, first2secondP)
                if rid2 == 0:
                    eresponse = " ".join([phrase_qualifiers[rid2],uresponsex])
                elif rid2 == 1:
                    eresponse = " ".join([phrase_qualifiers2a[0],phrase_qualifiers2a[randint(1,2)],uresponsex])
                elif  rid2 == 2:
                    eresponse = " ".join([phrase_qualifiers2b[0],phrase_qualifiers2b[randint(1,2)],uresponsex])
                print("\n")
                print_delayed(eresponse.strip().capitalize()+"?",1,"\n")
                
            ################################
            #Mode 1: heges (r1)
            ################################
            if rid == 1:
                print("\n")
                print_delayed(phrase_hedges[randint(1,2)].strip().capitalize(),1,"\n")
                for i in range(0,3):
                    print_delayed(".",0,'')
                print(' ',end='')
                print_delayed(phrase_hedges[0].strip().capitalize(),1,"\n")
                
            ################################
            #Mode 1: status (r2)
            ################################
            if rid == 2:
                print("\n")
                print_delayed(phrase_status[randint(0,1)].strip().capitalize(),1,"\n")
                for i in range(0,3):
                    print_delayed(".",1,'')
                print(' ',end='')
                
                if randint(0,1)==1:
                    if randint(0,1)==1:
                        print_delayed(phrase_atrisk[1],1,'\n')
                        uresponse = str(input(inputDialog2)).lower()
                        if uresponse == 'y':
                            atrisk = 1
                            break
                        print_delayed(phrase_atrisk[0],1,'\n')
                        mode = 2
                    else:
                        print_delayed(phrase_optout[randint(0,1)],2,'\n')
                        uresponse = str(input(inputDialog2)).lower()
                        if uresponse == 'y':
                            eresponse = " ".join(["Nice!", phrase_status[0],uresponsex])
                            print_delayed(eresponse[1],1,'\n')
                            mode = 2
                else:
                    print_delayed(phrase_optout2[0],1)
                    print_delayed(phrase_optout2[randint(1,3)],2)
            
            ################################
            #Mode 1: status (r4)
            ################################
            if rid == 4:
                print("\n")
                if randint(3,4)==3:
                    print_delayed(phrase_hedges[3].strip().capitalize(),1,"\n")
                    for i in range(0,3):
                        print_delayed(".",1,'')
                    print(' ',end='')
                    print_delayed(phrase_optout2[0],1)
                    print_delayed(phrase_optout2[randint(1,3)],2)
                else:
                    print_delayed(phrase_hedges[4].strip().capitalize(),1,"\n")
        
        uresponse = str(input(inputDialog)).lower()

    #####################
    #Execute Slow Exit
    #####################
    print("\n")
    if atrisk:
        print_delayed(phrase_ohdear2[0],1)
        print_delayed(phrase_ohdear2[1],1,'\n\n')
    else:
        print_delayed(phrase_ohdear1[0],0)
        for i in range(0,3):
            print_delayed(".",1,'')
        print(' ',end='')
        print_delayed(phrase_ohdear1[1],1,'\n\n')
        
    print_delayed(phrase_outro[0],2,'\n')
    print_delayed(phrase_outro[1],2,'\n')
    print_delayed(phrase_outro[2],2,'\n\n')
    for i in range(0,3):
        print_delayed(".",1,'')
    print(' ',end='')
    print_delayed(phrase_outro2[0],2)
    print_delayed(phrase_outro2[1],1)
    print("\n")
    print_delayed(phrase_positive[randint(0,1)],1)
```

```python
def main():
    elizaBot()
    
main()
```
