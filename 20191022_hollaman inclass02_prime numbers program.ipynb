# Exercise: Prime Numbers

Write an appliation that will that show a list of numbers and indicate whether or not they are prime.

The user will have to input the last number in the range, and you will print all of the numbers from 1 to that number.

```python
#hollaman
#20191022
#inclass02 Prime Numbers Exercise

#Introduction
print('Hey there! This is the Prime Number Printer. It\'ll print all the prime numbers from 1 to any integer you specify\n')

########################################################
#Part 1: get valid user input
########################################################

#set variable and while loop
response = 0
if response <= 0 or not isinstance(response, int):
    response = float(input('To begin, please enter an integer: '))
    
    #check for input error and rebroadcast
    while response <= 0 or not response.is_integer():
        response = float(input('Input Error. Please enter a positive integer to continue: '))
    
print('\nWonderful! You entered the number ' + str(int(response)) + '\nBelow we have printed all the primes numbers up to ' + str(int(response)) +" inclusive")


########################################################
#Part 2: check if input value is prime
########################################################
primes = set()
dubbs = set()

#for counter in range(1,response)
counter = 1
while counter <= response:
    evendivisors=0
    for x in range(1,counter+1,1): #choose 'counter+1' as the stop point since outputs for a range call will be less than the stated stop point and counter is our desired stop point and we are incrementing each step by 1
        
        ####################################################
        #approach#1 uses division + boolean isinteger check
        ####################################################
        divisor = float(counter/x) #iterative divide by a number between 1 and the counter, inclusive
        #print('divisor = ' + str(divisor))
        if divisor.is_integer(): #if result is integer, then it is counted as an 'even divisor'
            evendivisors+=1
        
        ####################################################    
        ##approach#2 uses modulo
        ####################################################
        #if divmod(counter,x)[0]>=1 and divmod(counter,x)[1]==0: #if result is integer, then it is counted as an 'even divisor'
        #    evendivisors+=1
            
    
    ####################################################    
    #some quick accounting + prime indicator
    ####################################################
    #print('evendivisors = ' + str(evendivisors))
    if evendivisors == 2: #the number being held in the counter variable is prime if it has only 2 event divisors
        primes.add(counter)
    else:
        dubbs.add(counter)
    
    #increment counter
    counter+=1


########################################################
#Part 3: printing results
########################################################
#first we make sure there are no intersections between the sets
if not primes&dubbs: #equivalent to: if the is the intersection of the sets is null
    print('####     PRIME NUMBERS     ####')
    for n in sorted(primes): #we sorted the set since in some executions that we ran the output of the for loops was not in ascending order
        print(str(n)+' is prime')

    print('\n####SOOOO NOT PRIME NUMBERS####')
    for m in sorted(dubbs):
        print(str(m)+' is sooo not prime')
else:
    print('There has been some computational error. Please consult (shake your fist at) the (very fallable) writer of this program to address')
```
